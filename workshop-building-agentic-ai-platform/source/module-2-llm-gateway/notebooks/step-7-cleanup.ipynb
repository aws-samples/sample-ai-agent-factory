# Module 2 -- Step 7: Cleanup

**Do not run cleanup now.**

Module 2 resources — virtual keys, teams, and especially the Bedrock Guardrail created in Step 5 — are consumed by later modules (3a, 3b, and 4). Deleting them here would break the downstream walkthroughs.

When you have finished the entire workshop, open the **Workshop Cleanup** page in the left-hand workshop navigation (Cleanup section). That page tears down every participant-created resource across all modules in the correct order.

The `workshop-llm-gateway-stack` itself is pre-provisioned by Workshop Studio and is torn down automatically when the workshop event ends — you do not delete it manually.
